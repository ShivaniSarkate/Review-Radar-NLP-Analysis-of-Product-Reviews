**project Title : ReviewRadar - NLP Analysis of Product Reviews**

**Objective** : Analyze customer reviews using NLP techniques to uncover sentiment, recurring themes, and actionable feedback for product improvement.

**Business Context** : Customer reviews contain rich insights about product strengths and weaknesses. NLP helps extract patterns that guide design, support, and marketing.

**Workflow Overview** :
- Simulate or import product reviews  
- Clean and preprocess text  
- Perform sentiment analysis  
- Extract keywords using TF-IDF  
- Apply topic modeling (LDA)  
- Visualize insights with word clouds and bar charts

```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from textblob import TextBlob
from wordcloud import WordCloud

# Step 2: Simulate reviews
reviews = [
    "The battery life is amazing and lasts all day.",
    "Camera quality is poor in low light.",
    "Fast performance but gets hot quickly.",
    "Excellent display and smooth touch response.",
    "Speaker volume is too low.",
    "Great value for money. Highly recommend!"
]
df = pd.DataFrame({'Review': reviews})

# Step 3: Sentiment analysis
df['Sentiment'] = df['Review'].apply(lambda x: TextBlob(x).sentiment.polarity)

# Step 4: TF-IDF and topic modeling
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(df['Review'])
lda = LatentDirichletAllocation(n_components=2, random_state=42)
lda.fit(X)

# Step 5: Word cloud
text = " ".join(df['Review'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Review Word Cloud')
plt.tight_layout()
plt.show()
```

**Executive Summary** :
- **Positive Themes**: Battery life, display, value for money  
- **Negative Themes**: Camera in low light, overheating, speaker volume  
- **Sentiment Score Range**: -0.3 to +0.8  
- **Recommendation**:  
  - Improve camera performance in low light  
  - Address thermal issues during heavy usage  
  - Highlight battery and display strengths in marketing  
- **Impact**: NLP reveals actionable product feedback and guides roadmap decisions
